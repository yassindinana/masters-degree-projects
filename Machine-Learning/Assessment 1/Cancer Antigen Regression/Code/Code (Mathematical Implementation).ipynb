```
def ridge(it , x , y , lr , alpha , m  ):
    w=np.zeros((1 , 10))
    for i in range(it):
        error=prd(w , x)-y
        w=w-lr*(1/m)*(np.dot( error.T , x )+alpha*(np.square(w)))
        print(w)
    return w
```
